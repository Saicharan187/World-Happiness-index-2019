# Building a Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X=df.iloc[:,2:].values
y=df.iloc[:,1:2].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
Regression=LinearRegression()
Regression.fit(X_train,y_train)
y_pred=Regression.predict(X_test)
mean_squared_error(y_test,y_pred)
```

```python
import statsmodels.formula.api as sm
from statsmodels.api import OLS
sm.OLS(y,X).fit().summary()
```

```python
Rd_State=[]
MSE=[]
for i in range(0,101,1):
    X=df.iloc[:,2:].values
    y=df.iloc[:,1:2].values
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=i)
    Regression=LinearRegression()
    Regression.fit(X_train,y_train)
    y_pred=Regression.predict(X_test)
    e=mean_squared_error(y_test,y_pred)
    Rd_State.append(i)
    MSE.append(e)
RdvsMSE=pd.DataFrame({'RD_State':Rd_State,'MSE':MSE})
RdvsMSE[RdvsMSE.MSE==RdvsMSE.MSE.min()]
#RdvsMSE
```

Lets Note that at Random State 80 We are getting Minimum amount of error

```python
Test=[]
MSE=[]
for i in np.arange(0.20,0.30,0.01):
    a=float("%.2f" % i)
    X=df.iloc[:,2:].values
    y=df.iloc[:,1:2].values
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=a,random_state=80)
    Regression=LinearRegression()
    Regression.fit(X_train,y_train)
    y_pred=Regression.predict(X_test)
    e=mean_squared_error(y_test,y_pred)
    Test.append(a)
    MSE.append(e)
TsvsMSE=pd.DataFrame({'TestSize':Test,'MSE':MSE})
TsvsMSE[TsvsMSE.MSE==TsvsMSE.MSE.min()]
```

Lets make a note that At Test size 0.29 we are getting minimum error 0.156

```python
X=df.iloc[:,2:].values
y=df.iloc[:,1:2].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.29,random_state=80)
Regression=LinearRegression()
Regression.fit(X_train,y_train)
y_pred=Regression.predict(X_test)
mean_squared_error(y_test,y_pred)
```

Let's look another way Feature Reduction process to build best fit line and minimize the error

```python
X=df.iloc[:,3:].values
y=df.iloc[:,1:2].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.29,random_state=80)
Regression=LinearRegression()
Regression.fit(X_train,y_train)
y_pred=Regression.predict(X_test)
mean_squared_error(y_test,y_pred)
```

We tried Feature Reduction process here so that best fit line can have lower error points. From The OLS Summary we seen Variable
GDP per Capita has lesser Co efficient so we removed that variable and we jumped from 0.156 Error to 0.126 Error
